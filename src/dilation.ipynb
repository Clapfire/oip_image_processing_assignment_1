# Dilation

This module will demonstrate how to dilate an image. The implementation is not very efficient, but implements formal dilation.

```python
# Import the necessary libraries.
import modules.oiplib as oiplib
import numpy as np
import time
```

```python
# Load the image.
grayImg = oiplib.rgb2GrayLuminosity(oiplib.loadImage('../images/heptamers.png'))
oiplib.showImage(grayImg, "Original image", figsize=(10, 10));
```

```python
# Convert grayscale image to binary image.
binImg = oiplib.gray2Binary(grayImg)
oiplib.showImage(binImg, "Binary image", figsize=(10, 10), vmax=1);
```

```python
# Examples for how to generate explicit structuring elements for a given center coordinate.
se0 = oiplib.createStructuringElement(1, "8N")((4, 4))
oiplib.showImage(oiplib.set2Binary(se0), "Neighborhood image (r = 1, nbh = 8N)", figsize=(5, 5), vmax=1);

se1 = oiplib.createStructuringElement(1, "4N")((4, 4))
oiplib.showImage(oiplib.set2Binary(se1), "Neighborhood image (r = 1, nbh = 4N)", figsize=(5, 5), vmax=1);

se2 = oiplib.createStructuringElement(2, "8N")((4, 4))
oiplib.showImage(oiplib.set2Binary(se2), "Neighborhood image (r = 2, nbh = 8N)", figsize=(5, 5), vmax=1);

se3 = oiplib.createStructuringElement(2, "4N")((4, 4))
oiplib.showImage(oiplib.set2Binary(se3), "Neighborhood image (r = 2, nbh = 4N)", figsize=(5, 5), vmax=1);

se2 = oiplib.createStructuringElement(3, "8N")((4, 4))
oiplib.showImage(oiplib.set2Binary(se2), "Neighborhood image (r = 3, nbh = 8N)", figsize=(5, 5), vmax=1);

se3 = oiplib.createStructuringElement(3, "4N")((4, 4))
oiplib.showImage(oiplib.set2Binary(se3), "Neighborhood image (r = 3, nbh = 4N)", figsize=(5, 5), vmax=1);
```

```python
# Start performance timer.
start = time.perf_counter()

# Convert binary image to set.
setImg = oiplib.binary2Set(binImg)

# Dilate set.
dilatedSetImg = oiplib.dilateSet(setImg, oiplib.createStructuringElement(5))

# Convert set image back to binary.
dilatedBinImg = oiplib.set2Binary(dilatedSetImg, binImg.shape)

# Stop performance timer.
stop = time.perf_counter()

# Show dilated picture.
oiplib.showImage(dilatedBinImg, "Dilated image", figsize=(10, 10), vmax=1);
```

```python
# Display runtime performance.
print("Time elapsed during dilation: %.2fs"%(stop - start))
```
