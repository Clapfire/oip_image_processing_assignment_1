# DPI (Resolution detection)

This module will demonstrate how to determine the resolution from the image legend. The pipeline performs the following operations:
    
    - Load image.
    - Convert image to grayscale.
    - Extract scan line from legend.
    - Get first and last foreground pixel and subtract indices.
    - Use length to calculate m/px.

```python
# Import the image processing library.
import modules.oiplib as oiplib
import numpy as np
```

```python
# Load the image.
grayImg = oiplib.rgb2GrayLuminosity(oiplib.loadImage('../images/heptamers_legend.png'))
oiplib.showImage(grayImg, "Original image", figsize=(10, 10));
```

```python
# Calculate the size of a pixel in meters.
pixelSize = oiplib.measurePixelSize(grayImg)
```

```python
# Display pixel size.
print("One pixel has the size of %.4f nm in the real world."%(pixelSize*1e9))
```
